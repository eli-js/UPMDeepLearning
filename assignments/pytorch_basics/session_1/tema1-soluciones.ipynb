### Ejercicio 18: Crea una función para multiplicar dos tensores. Si las dimensiones son incompatibles debe devolver un mensaje de error.

```python
def multiply_tensors_1(tensor1, tensor2):
    if tensor1.shape != tensor2.shape:
        raise ValueError("Both tensors must have the same shape to multiply.")

    return tensor1*tensor2

def multiply_tensors_2(tensor1, tensor2):
    if tensor1.shape != tensor2.shape:
        raise ValueError("Both tensors must have the same shape to multiply.")

    return torch.multiply(tensor1, tensor2)
```

### Ejercicio 34: Crea una función que dado un tensor de 2 dimensiones y otro de 1 dimensión, los sume.

Ejemplo:
\begin{bmatrix}
0.5 & 0.7 & 0.2 \\
0.6 & 0.1 & 0.4 \\
0.3 & 0.8 & 0.9 \\
\end{bmatrix}
<h3 align="center">+</h3>
\begin{bmatrix}
10 & 20 & 30
\end{bmatrix}
<h3 align="center">=</h3>
\begin{bmatrix}
10.5 & 20.7 & 30.2 \\
10.6 & 20.1 & 30.4 \\

```python
from torch.nn import Parameter
```

```python
class LinearLayer(torch.nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LinearLayer, self).__init__()

        self.weight = Parameter(torch.randn(output_dim, input_dim))
        self.bias = Parameter(torch.randn(output_dim))

    def forward(self, x):

        return x @ self.weight.T + self.bias
```
